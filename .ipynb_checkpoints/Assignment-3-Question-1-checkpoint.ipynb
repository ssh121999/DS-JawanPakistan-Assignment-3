```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("netflix_data.csv")
```

```python
data.head()
```

```python
data["cast"] = data["cast"].replace(np.nan, "None")
```

```python
data.head(n=30)
```

```python
def count_cast(cast):
    if cast == "None":
        return 0
    else:
        arr_length = cast.split(", ")
        return len(arr_length)

data["num_of_cast"] = data["cast"].apply(count_cast)
data.head(n=30)
```
